```python
import requests
import bs4 as bs

resp = requests.get('http://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
soup = bs.BeautifulSoup(resp.text, 'lxml')
table = soup.find('table', {'class': 'wikitable sortable'})

tickers = []
names = []
for row in table.findAll('tr')[1:]:
    ticker = row.findAll('td')[0].text.replace("\n", "")
    name = row.findAll('td')[1].text.replace("\n", "")
    tickers.append(ticker)
    names.append(name)
```

```python
len(tickers)
```

```python
import pandas as pd
companies = pd.read_csv('companies.csv')
```

```python
companies.head(1)
```

```python
SnP500 = pd.DataFrame({'ticker':tickers,'name':names})
```

```python
SnP500
```

```python
def func(name,names):
    if name in names:
        return 1
    else:
        return 0
```

```python
names = companies['name'].to_list()
```

```python
SnP500['bool'] = SnP500.apply(lambda x : func(x['name'],names),axis=1)
```

```python
result = SnP500[SnP500['bool'] == 1]
result.head(10)
```

```python
tickers = result['ticker'].to_list()
```

```python
import yfinance as yt

ticker_info_lst = list(map(lambda t: yt.Ticker(t).info, tickers))
print('{} S&P 500 stocks for this example are: {}'.format(len(ticker_info_lst), tickers))
```

### Recale the data to put them into spider plot

```python
import pandas as pd
```

```python
df = pd.read_csv('SpiderChart.csv',index_col=0)
```

```python
df.head()
```

```python
df.columns
```

```python
useful_columns = ['shortName','fullTimeEmployees','volume','currentRatio','bookValue','totalCashPerShare','currentPrice','grossMargins']
```

```python
result = df[useful_columns]
```

```python
result.head()
```

```python
def delete_marks(str):
    return str.replace(',',';')
```

```python
result['shortName'] = result.apply(lambda x: delete_marks(x['shortName']),axis=1)
```

```python
result[useful_columns[1]]=result[useful_columns[1]]/100000
result[useful_columns[2]]=result[useful_columns[2]]/2000000
result[useful_columns[3]]=result[useful_columns[3]]/2
result[useful_columns[4]]=result[useful_columns[4]]/100
result[useful_columns[5]]=result[useful_columns[5]]/10
result[useful_columns[6]]=result[useful_columns[6]]/200
result[useful_columns[7]]=result[useful_columns[7]]*2
```
